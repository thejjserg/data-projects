## Turn Up the Temperature

#### Task 1
Write a function called `f_to_c` that takes an input `f_temp`, a temperature in Fahrenheit, and converts it to `c_temp`, the temperature in Celsius.

It should then return `c_temp`.

The equation you should use is:
```
Temp (C) = (Temp (F) - 32) * 5/9
```

````python
# Task 1
def f_to_c(f_temp):
    return (f_temp - 32) * 5 / 9
````

#### Task 2

Let’s test your function with a value of `100` degrees Fahrenheit.

Define a variable `f100_in_celsius` and set it equal to the value of `f_to_c()` with `100` as the input.

````python
# Task 2
f100_in_celsius = f_to_c(100)

# Test for Task 2
#print(f100_in_celsius)
````

#### Task 3

Write a function called `c_to_f` that takes an input `c_temp`, a temperature in Celsius, and converts it to `f_temp`, the temperature in Fahrenheit.

It should then return `f_temp`.

The equation you should use is:
```
Temp (F) = Temp (C) * (9/5) + 32
```

````python
# Task 3
def c_to_f(c_temp):
    return c_temp * (9/5) + 32
````

#### Task 4

````python
# Task 4
c0_in_fahrenheit = c_to_f(0)

# Test for Task 4
#print(c0_in_fahrenheit)
````

## Use the Force

#### Task 5
Define a function called `get_force` that takes in `mass` and `acceleration`. It should return `mass` multiplied by `acceleration`.

````python
# Task 5
def get_force(mass, acceleration):
    return mass * acceleration
````

#### Task 8
Define a function called `get_energy` that takes in `mass` and `c`.

`c` is a constant that is usually set to the speed of light, which is roughly `3 x 10^8`. Set `c` to have a default value of `3*10**8`.

`get_energy()` should return `mass` multiplied by `c` squared.

````python
# Task 8
def get_energy(mass, c = 3*10**8):
    return mass * c**2
````

#### Task 9

````python
# Task 9
bomb_energy = get_energy(bomb_mass)
````

## Do the Work
#### Task 11
Define a final function called `get_work` that takes in `mass`, `acceleration`, and `distance`.

Work is defined as `force` multiplied by `distance`. First, get the `force` using `get_force()`, then multiply that by `distance`. Return the result.

````python
# Task 11
def get_work(mass, acceleration, distance):
    force = get_force(mass = mass, acceleration = acceleration)

    return force * distance
````

#### Task 12

Test `get_work()` by using it with `train_mass`, `train_acceleration`, and `train_distance`. Save the result to a variable called `train_work`.

````python
# Task 12
train_work = get_work(train_mass, train_acceleration, train_distance)
````

#### Task 13
Print the string `The GE train does X Joules of work over Y meters.`, with `X` replaced with `train_work` and `Y` replaced with `train_distance`.

````python
# Task 13
print("The GE train does " + str(train_work) + " Joules of work over " + str(train_distance) + " meters.")
````

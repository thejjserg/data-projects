## Solution:

#### Task 8
Great job! Now, test everything one more time!

What is the cheapest method of shipping a 4.8 pound package and how much would it cost?

What is the cheapest method of shipping a 41.5 pound package and how much would it cost?


````python
# Task 8
def optimize(weight):
    # Calculate the various options
    ground = ground_shipping(weight)
    premium = ground_premium_shipping()
    drone = drone_shipping(weight)

    # Add a New Line
    print("")

    # Bring the items into a list
    options = [ground, premium, drone]

    # Find the Cheapest Option
    cheapest = min(options)

    if cheapest == ground:
        print("The cheapest method for shipping is Ground Shipping.")

    elif cheapest == premium:
        print("The cheapest method for shipping is Premium Ground Shipping.")

    elif cheapest == drone:
        print("The cheapest method for shipping is Drone Shipping.")

    else:
        print("There was an error.")
        return 0
    
    print("The cost for this method is $" + str(cheapest) + " for a " + str(weight) + "lb package.")

    return cheapest
````
